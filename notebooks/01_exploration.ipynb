```python
LAP_COL = "lap"
CAR_COL = "vehicle_id"
TIME_COL = "value"

lap_basic = lap_times_r1[[CAR_COL, LAP_COL, TIME_COL]].dropna().copy()

lap_basic.rename(columns={
    CAR_COL: "car_id",
    LAP_COL: "lap_number",
    TIME_COL: "lap_time_s",
}, inplace=True)

lap_basic.head()
```

```python
# Median lap pace per car
lap_basic["median_pace"] = (
    lap_basic.groupby("car_id")["lap_time_s"].transform("median")
)

# Label outliers (slow laps)
lap_basic["pit_like"] = lap_basic["lap_time_s"] > lap_basic["median_pace"] * 1.35

lap_basic.head()
```

```python
lap_basic = lap_basic.sort_values(["car_id", "lap_number"])

lap_basic["cumulative_time_s"] = (
    lap_basic.groupby("car_id")["lap_time_s"].cumsum()
)
```

```python
race_state = []

for lap in sorted(lap_basic["lap_number"].unique()):
    subset = lap_basic[lap_basic["lap_number"] == lap].copy()
    subset = subset.sort_values("cumulative_time_s")
    
    subset["position"] = range(1, len(subset) + 1)
    
    leader_time = subset["cumulative_time_s"].iloc[0]
    subset["gap_to_leader_s"] = subset["cumulative_time_s"] - leader_time
    
    subset["gap_to_front_s"] = subset["cumulative_time_s"].diff().fillna(0)
    
    race_state.append(subset)

race_state_df = pd.concat(race_state).reset_index(drop=True)

race_state_df.head(20)
```

```python
import pandas as pd

# Load start/end datasets
lap_start = pd.read_csv(race_1_path / "road_america_lap_start_R1.csv")
lap_end   = pd.read_csv(race_1_path / "road_america_lap_end_R1.csv")

print(lap_start.columns)
print(lap_end.columns)

lap_start.head(), lap_end.head()
```

```python
# Rename columns for clarity
lap_start = lap_start.rename(columns={"value": "start_raw"})
lap_end   = lap_end.rename(columns={"value": "end_raw"})

# Convert timestamps if needed
lap_start["timestamp"] = pd.to_datetime(lap_start["timestamp"])
lap_end["timestamp"]   = pd.to_datetime(lap_end["timestamp"])

# Merge
laps = lap_start.merge(lap_end, on=["lap", "vehicle_id"], suffixes=("_start", "_end"))

# Compute lap time
laps["lap_time_s"] = (laps["timestamp_end"] - laps["timestamp_start"]).dt.total_seconds()

# Preview
laps.head()
```

```python
import pandas as pd

# if your current df is named something else, replace `laps` below
race_laps = laps.copy()

# Standardize column names
race_laps = race_laps.rename(columns={
    "vehicle_id": "car_id",
    "lap": "lap_number",
    # "lap_time_s" is already fine
})

# Optional: if lap 0 is formation/grid lap and you want only race laps:
race_laps = race_laps[race_laps["lap_number"] > 0]
```

```python
race_laps = race_laps.sort_values(["car_id", "lap_number"])

race_laps["cumulative_time_s"] = (
    race_laps.groupby("car_id")["lap_time_s"].cumsum()
)

race_laps.head()
```

```python
race_laps["median_pace"] = (
    race_laps.groupby("car_id")["lap_time_s"].transform("median")
)

race_laps["pit_like"] = race_laps["lap_time_s"] > race_laps["median_pace"] * 1.35
```

```python
# Sort by lap then by cumulative time (who reached that lap first)
race_state_df = race_laps.sort_values(
    ["lap_number", "cumulative_time_s"]
).copy()

# Position within each lap: 1, 2, 3, ...
race_state_df["position"] = (
    race_state_df.groupby("lap_number")["cumulative_time_s"]
    .rank(method="first")
    .astype(int)
)

# Gap to leader at each lap
race_state_df["gap_to_leader_s"] = (
    race_state_df["cumulative_time_s"]
    - race_state_df.groupby("lap_number")["cumulative_time_s"].transform("min")
)

# Gap to car directly in front at that lap
race_state_df["gap_to_front_s"] = (
    race_state_df.groupby("lap_number")["cumulative_time_s"].diff().fillna(0)
)

race_state_df.head(20)
```

```python
import pandas as pd

# Start from your current `laps` df
# Keep only what we actually need for race logic
race_laps = laps[["vehicle_id", "lap", "lap_time_s"]].copy()

# Rename to standard names
race_laps.rename(columns={
    "vehicle_id": "car_id",
    "lap": "lap_number"
}, inplace=True)

# Clean types
race_laps = race_laps.dropna(subset=["lap_time_s"])
race_laps["lap_number"] = race_laps["lap_number"].astype(int)
race_laps["lap_time_s"] = race_laps["lap_time_s"].astype(float)

# Sort by car, then lap
race_laps = race_laps.sort_values(["car_id", "lap_number"])
race_laps.head()
```

```python
# Cumulative race time per car
race_laps["cumulative_time_s"] = (
    race_laps.groupby("car_id")["lap_time_s"].cumsum()
)

# Median pace per car
race_laps["median_pace"] = (
    race_laps.groupby("car_id")["lap_time_s"].transform("median")
)

# Simple pit / slow-lap flag
race_laps["pit_like"] = race_laps["lap_time_s"] > race_laps["median_pace"] * 1.35

race_laps.head()
```

```python
# Sort by lap & cumulative time (who reached that lap first)
race_state_df = race_laps.sort_values(
    ["lap_number", "cumulative_time_s"]
).copy()

# 1) Position on each lap
race_state_df["position"] = (
    race_state_df.groupby("lap_number")["cumulative_time_s"]
    .rank(method="first")
    .astype(int)
)

# 2) Gap to leader at that lap
race_state_df["gap_to_leader_s"] = (
    race_state_df["cumulative_time_s"]
    - race_state_df.groupby("lap_number")["cumulative_time_s"].transform("min")
)

# 3) Gap to car directly in front
race_state_df["gap_to_front_s"] = (
    race_state_df.groupby("lap_number")["cumulative_time_s"].diff().fillna(0)
)

race_state_df = race_state_df.reset_index(drop=True)
race_state_df.head(15)
```
